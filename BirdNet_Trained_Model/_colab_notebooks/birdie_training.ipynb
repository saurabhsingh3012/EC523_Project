**READ ME:**

This script leverages the most recenve version of the BirdNet-Analyzer V2.3 published April 2023: https://github.com/kahst/BirdNET-Analyzer

Prior versions of Bird-Net analyzer cannot be cannot be run on SCC due to OS legacy compatibility issues with glibc. Re INC20042608 with BU IT for additional details. We spent weeks trying to get this to run. 


Steps:
1. Clone latest version of Bird-Net Repository https://github.com/kahst/BirdNET-Analyzer to your GDrive or AWS environment
2. Place this script into the same location as the 'model.py', 'analyze.py' files
3. Download the Data Set from our team GIT repository
3. Update the file path locations to injest the data
4. To train, modify the train.py script. Note that many tuning parameters can be passed in as arugments. If you want to modify the layer parameters, then you will need to modify the model.py and train.py scripts.
5. To analyze, run the analyze.py script. Make sure to specify the location of your trained Classifiers!
